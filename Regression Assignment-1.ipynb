### Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?

#### Assumptions of linear regression:

* **Linearity** : The relationship between the independent variables and the dependent variable should be a straight line.
    * Check by plotting the dependent variable against each independent variable and see if the points form a roughly straight line.

* **Independence** : Each observation in the dataset should be independent and not influenced by other observations.
    * Ensure that the values of the dependent variable for different observations are not related or affected by each other.
    
* **Homoscedasticity** : The spread of the residuals (the differences between observed and predicted values) should be similar across all levels of the independent variables.
   * Check by plotting the residuals against the predicted values and see if the points are scattered randomly around zero without any pattern.
   
* **Normality** : The residuals should follow a normal distribution, meaning they resemble a bell-shaped curve.
  * Use statistical tests or visual methods like histograms or Q-Q plots to assess if the residuals look like a bell curve.
  
* **No multicollinearity** : The independent variables should not be highly correlated with each other.
  * Calculate the variance inflation factor (VIF) for each independent variable. Values above 5 or 10 may indicate high multicollinearity

**Slope:** The slope represents how the price of a pizza changes for each additional inch in size. It tells us the average increase or decrease in price per inch.

For example, if the slope is $2, we can interpret it as follows:

For every additional inch in pizza size, the price is estimated to increase by $2.

So, in this simple scenario, the interpretation is straightforward: Each extra inch in pizza size corresponds to a $2 increase in price, assuming there are no other factors considered in the model.

### Q4. Explain the concept of gradient descent. How is it used in machine learning?

Gradient descent is an optimization algorithm used in machine learning to minimize the error or cost function of a model. It is widely used in training various types of machine learning models, including linear regression, logistic regression, and neural networks.

The concept of gradient descent revolves around finding the minimum of a function by iteratively updating the parameters or weights of the model. In the context of machine learning, the goal is to find the optimal set of parameters that minimize the difference between the predicted output and the actual output.

The key idea behind gradient descent is that by iteratively adjusting the parameters in the direction of steepest descent, the algorithm can converge to the optimal parameter values that minimize the cost function. The gradient provides information about the slope or direction of the function at a specific point, guiding the algorithm towards the minimum.

### Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?

Multiple linear regression is an extension of simple linear regression that allows for the modeling of the relationship between a dependent variable and multiple independent variables. It assumes a linear relationship between the dependent variable and the independent variables, but with the inclusion of multiple predictors.

In simple linear regression, we have a single independent variable (predictor) and a dependent variable. The goal is to find a linear relationship between the predictor and the dependent variable, represented by a straight line equation (y = mx + c). The model estimates the slope (m) and the intercept (c) that minimize the sum of squared differences between the predicted and actual values.

On the other hand, multiple linear regression involves two or more independent variables (predictors) and a single dependent variable. The goal is to find a linear equation that best represents the relationship between the predictors and the dependent variable.

The multiple linear regression model can be represented by the equation:

y = b0 + b1x1 + b2x2 + ... + bnxn + ε

where:
- y is the dependent variable
- b0 is the intercept or constant term
- b1, b2, ..., bn are the coefficients corresponding to each independent variable (x1, x2, ..., xn)
- x1, x2, ..., xn are the independent variables
- ε represents the error term or residuals, accounting for the part of the dependent variable that is not explained by the predictors

The coefficients (b1, b2, ..., bn) represent the effect of each independent variable on the dependent variable while holding other variables constant. They are estimated using a statistical method called ordinary least squares (OLS), which minimizes the sum of squared residuals.

In multiple linear regression, the model can capture more complex relationships between the dependent variable and multiple predictors. It allows for the assessment of the individual contributions and significance of each predictor while considering the influence of other variables.

Additionally, multiple linear regression provides a way to analyze the combined effect of multiple predictors on the dependent variable, taking into account their interactions and dependencies. This makes it a powerful tool in various fields, such as economics, social sciences, and business analytics, where multiple factors influence an outcome.

It is worth noting that multiple linear regression assumes certain assumptions, including linearity, independence of errors, constant variance of errors (homoscedasticity), and absence of multicollinearity among predictors, among others. Violations of these assumptions can affect the reliability and interpretability of the regression results.

### Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?

Multicollinearity refers to a situation in multiple linear regression where two or more independent variables are highly correlated with each other. It can pose challenges in interpreting the regression model and estimating the individual effects of the correlated variables on the dependent variable.

Detecting multicollinearity:
1. Correlation Matrix: Calculate the correlation coefficients between all pairs of independent variables. High correlation coefficients (close to 1 or -1) indicate potential multicollinearity.

2. Variance Inflation Factor (VIF): Calculate the VIF for each independent variable. VIF measures how much the variance of the estimated regression coefficient is increased due to multicollinearity. VIF values greater than 1 indicate multicollinearity, and values above 5 or 10 are often considered problematic.

Addressing multicollinearity:
1. Remove a Variable: If two or more variables are highly correlated, consider removing one of them from the regression model. Choose the variable based on its relevance to the problem, domain knowledge, or theoretical importance.

2. Principal Component Analysis (PCA): PCA can be used to transform the original variables into a new set of uncorrelated variables (principal components). These components capture most of the variation in the data while reducing the impact of multicollinearity.

3. Domain Knowledge: Incorporate domain knowledge to interpret the variables and their relationships. Sometimes, even if variables are highly correlated, their inclusion might be necessary to capture important aspects of the problem.

4. Collect More Data: Increasing the sample size can help reduce the impact of multicollinearity by providing more information and reducing the uncertainty in estimating regression coefficients.

It's important to note that while addressing multicollinearity, the goal is to improve the interpretability and stability of the model without compromising its predictive performance.

### Q7. Describe the polynomial regression model. How is it different from linear regression?

Polynomial regression is a type of regression analysis that allows for modeling non-linear relationships between the dependent variable and the independent variables. Unlike linear regression, which assumes a linear relationship, polynomial regression captures non-linear patterns by including higher-order terms of the independent variable(s) in the model.

In linear regression, the relationship between the dependent variable and the independent variable(s) is represented by a straight line. The equation takes the form:

y = b0 + b1x1 + b2x2 + ... + bnxn + ε

where y is the dependent variable, x1, x2, ..., xn are the independent variables, b0, b1, b2, ..., bn are the coefficients, and ε is the error term.

In polynomial regression, the model includes not only the independent variables but also their higher-order terms. For example, in a polynomial regression model of degree 2, the equation takes the form:

y = b0 + b1x1 + b2x1^2 + b3x2 + b4x2^2 + ... + bnxn + ε

Here, x1^2, x2^2, etc., are the squared terms of the independent variables. By including these higher-order terms, polynomial regression can capture non-linear patterns and better fit the data.

The key difference between linear regression and polynomial regression is the flexibility in capturing non-linear relationships. Linear regression assumes a linear relationship, while polynomial regression allows for curves, bends, and other non-linear patterns in the data.

Overall, polynomial regression provides a way to model non-linear relationships between variables, offering more flexibility in capturing complex patterns in the data compared to linear regression.
