### Q1. Explain the concept of R-squared in linear regression models. How is it calculated, and what does it represent?

For Performance Metrics in regression there are 2 techniques used:

    1] R Squared 
    2] Adjusted R Squared
    
##### * R Squared



#### Formula :
         R Squared = 1 - SS(res) / SS(tot)
Where,
* SS(res) is Sum of Square residual / Error (sum of difference between actual and predicted point sqaured)
* SS(tot) is the Sum of square total 

Note: SS(res) > SS(total)


**R Squared Ranges between 0 to 1** 





###  Q2. Define adjusted R-squared and explain how it differs from the regular R-squared.

To understand the difference between R Squared and Adjusted R Squared, Let's consider a dataset which has the following features:

|Features|R Squared|Adjusted R Squared| Relation|
|--------|---------|------------------|---------|
|Size of House, Rooms & Price | 0.85 |0.79|All the features are Related|
|Size of House, Rooms,Location & Price| 0.90|0.85|All the features are related|
|Size of House, Rooms,Location,Gender & Price| 0.93|0.80|All except Gender are related|

As we can notice that the R sqaured value keeps on increasing with an increase in  the no. of features and the amount/quantity of increase in R Squared value is directly relational ot the relatability of new feature added.

While, in case of Adjusted R Squared, the value increases if there is a direct relation between the a new feature and Output feature. However, it decreases when there is no direct relation between the new feature and output feature.

### Q4. What are RMSE, MSE, and MAE in the context of regression analysis? How are these metrics calculated, and what do they represent?

RMSE, MSE, and MAE are cost functions in the context of regression analysis. They are used to measure the error.
RMSE is root mean squared error, MSE is mean squared error and MAE is mean absolute error.

## Mean Absolute Error (MAE)

RMSE, MSE, and MAE are all metrics used to evaluate the performance of a regression model. They all measure the difference between the predicted values and the actual values, but they do so in different ways.

* **RMSE** (root mean squared error) is the square root of the mean squared error. It is calculated by taking the average of the squared differences between the predicted values and the actual values. RMSE is a more sensitive measure of error than MAE, because it penalizes larger errors more heavily.
* **MSE** (mean squared error) is the average of the squared differences between the predicted values and the actual values. MSE is less sensitive to outliers than MAE, because it does not take the absolute value of the errors.
* **MAE** (mean absolute error) is the average of the absolute differences between the predicted values and the actual values. MAE is more robust to outliers than MSE, because it does not take the square of the errors.

In general, a lower value of RMSE, MSE, or MAE indicates a better performing regression model. However, the choice of which metric to use depends on the specific application. For example, if the data contains outliers, then MAE may be a better choice than MSE. If the data is not normally distributed, then RMSE may be a better choice than MAE.

Here are some examples of how these metrics can be used:

* A company is trying to predict the sales of a new product. They use a regression model to predict the sales, and they calculate the RMSE, MSE, and MAE. The RMSE is 10, the MSE is 20, and the MAE is 5. This tells the company that the average error in their predictions is 5, and that the largest error is 10.
* A doctor is trying to predict the risk of heart disease for a patient. They use a regression model to predict the risk, and they calculate the RMSE, MSE, and MAE. The RMSE is 0.1, the MSE is 0.2, and the MAE is 0.05. This tells the doctor that the average error in their predictions is 0.05, and that the largest error is 0.1.

These are just two examples of how RMSE, MSE, and MAE can be used to evaluate the performance of a regression model.
